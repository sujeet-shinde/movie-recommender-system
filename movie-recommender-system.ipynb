```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
movies.info()
```

```python
# Features selected for the project:
# - genres    : List of genres for each movie
# - id        : Unique movie identifier
# - keywords  : Keywords describing the movie content
# - title     : Title of the movie
# - overview  : Short description or synopsis of the movie
# - cast      : Main actors/actresses involved in the movie
# - crew      : Crew members (director, writer, etc.)
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
# Checking for missing (null) values in the dataset
movies.isnull().sum()
```

```python
# Dropping rows with null values
movies.dropna(inplace=True)
```

```python
# Checking for duplicate rows in the dataset
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# Converting the stringified list of dictionaries into a list of genre names
import ast
# Using ast.literal_eval() to safely convert the stringified list of dictionaries into a Python list of dictionaries

def convert(obj):
    list = []
    for i in ast.literal_eval(obj):
        list.append(i['name'])
    return list
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
# Same with the keywords
movies['keywords'].apply(convert)
```

```python
movies ['keywords'] = movies ['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convertcast(obj):
    list = []
    cnt = 0
    for i in ast.literal_eval(obj):
        if cnt != 3:
            list.append(i['name'])
            cnt+=1
        else:
            break
    return list
```

```python
# Extracting the names of the top 3 cast members from the stringified list of dictionaries in the 'cast' column

movies['cast'].apply(convertcast)
```

```python
movies['cast'] = movies['cast'].apply(convertcast)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    list = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            list.append(i['name'])
            break
    return list
```

```python
# Extracting the director's name from the stringified list of dictionaries in the 'crew' column

movies['crew'].apply(fetch_director)
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
# Converting the movie overview (a string) into a list of words for easier text processing

movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# Removing spaces from multi-word entries (like 'Science Fiction' â†’ 'ScienceFiction')
# so that each tag becomes a single word without spaces

movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']

movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
# Converting the list of tags into a single string so it's ready for text-based processing

new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
# Applying stemming to the tags column to reduce words to their root/base form
# Stemming helps normalize words that have the same meaning but different forms
# (e.g., "loved", "loving", "loves" â†’ "love"), reducing redundancy in textual data.
# This improves the efficiency of text vectorization and helps the model generalize better.
# We are using the PorterStemmer from NLTK, which is a commonly used stemming algorithm.

import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    lst = []
    for i in text.split():
        lst.append(ps.stem(i))

    return " ".join(lst)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# converting into lower case
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
# Applying the Bag of Words model to convert text data (tags) into numerical vectors for machine learning
# This will help quantify the textual information using word frequency
# Using CountVectorizer from scikit-learn with a max of 5000 features and removing English stop words

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
for word in cv.get_feature_names_out():
    print(word)
```

```python
# Calculating the cosine similarity between all movie vectors.
# Cosine similarity measures the angle between two vectors in multi-dimensional space.
# A smaller angle (i.e., higher cosine value close to 1) means the movies are more similar.
# This is useful for building a recommendation system where we find movies similar to a given one.

from sklearn.metrics.pairwise  import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
# Check similarity of 1st movie with itself (index 0)
print(similarity[0][0])
```

```python
# Check similarity of 2nd movie with itself (index 1)
print(similarity[1][1])
```

```python
# Creating a list of (index, similarity_score) pairs for the first movie (index 0)
# Using enumerate allows us to retain the original indices of movies before sorting
# This is important because we'll use these indices later to fetch the actual movie titles in the recommend function

list(enumerate(similarity[0]))
```

```python
# Sorting the list of (movie_index, similarity_score) tuples in descending order of similarity
# Using key=lambda x: x[1] to sort based on similarity score (not index)
# [1:6] is used to skip the first result (the movie itself with similarity 1.0) and fetch the top 5 most similar movies

sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
# Testing the recommendation model to verify if it returns output correctly
# This helps ensure the logic works before final integration or UI development
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```

```python
recommend('John Carter')
```
